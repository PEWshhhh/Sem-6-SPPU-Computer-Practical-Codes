```python
import nltk
import pandas as pd

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Download required NLTK resources

nltk.download('punkt')

nltk.download('punkt_tab')

nltk.download('stopwords')

nltk.download('averaged_perceptron_tagger')

nltk.download('averaged_perceptron_tagger_eng')

nltk.download('wordnet')
```

```python
# Sample document

document = """
Today is the end of the Republic,
The end of a regime that acquiesces to disorder.
At this very moment, in a system far from here, 
the New Republic lies to the galaxy while secretly
supporting the treachery.
"""

print(document)
```

```python
# Tokenization

tokens = word_tokenize(document)

print("Tokens :")

print(tokens)
```

```python
# POS Tagging

pos_tags = pos_tag(tokens)

print("POS Tags :")

print(pos_tags)
```

```python
# Stop Words Removal

stop_words = set(stopwords.words('english'))

filtered_words = [
    word for word in tokens
    if word.lower() not in stop_words
]

print("Words after Stop Words Removal :")

print(filtered_words)
```

```python
# Stemming

stemmer = PorterStemmer()

stemmed_words = [
    stemmer.stem(word)
    for word in filtered_words
]

print("Stemmed Words :")

print(stemmed_words)
```

```python
# Lemmatization

lemmatizer = WordNetLemmatizer()

lemmatized_words = [
    lemmatizer.lemmatize(word)
    for word in filtered_words
]

print("Lemmatized Words :")

print(lemmatized_words)
```

```python
# Create sample documents

documents = [
    "Text analytics is very useful",
    "Natural language processing is important",
    "Text mining and text analytics are related"
]

print(documents)
```

```python
# TF-IDF Vectorization

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(documents)
```

```python
# Display feature names

feature_names = vectorizer.get_feature_names_out()

print(feature_names)
```

```python
# Convert TF-IDF matrix into dataframe

tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=feature_names
)

tfidf_df
```
